```python
import tensorflow as tf
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

TensorFlow has several ops that create random tensors with different distributions. The random ops are stateful, and create new random values each time they are evaluated.

The seed keyword argument in these functions acts in conjunction with the graph-level random seed. Changing either the graph-level seed using set_random_seed or the op-level seed will change the underlying seed of these operations. Setting neither graph-level nor op-level seed, results in a random seed for all operations. See set_random_seed for details on the interaction between operation-level and graph-level random seeds.

------

## tf.random_normal(shape, mean=0.0, stddev=1.0, dtype=tf.float32, seed=None, name=None)

Возвращает тензор случайных значений заданной формы нормального распределения

Например:

```python
a = tf.random_normal([10],0,1)
```

```python
with tf.Session() as sess:
    y = sess.run(a)
    x = sess.run(tf.range(10))
    print(y)
    print(x)
```

Визуально это выглядит вот так:

```python
plt.figure(figsize=(10,10))

mean, = plt.plot(x,y,color='green',linestyle='--')
original, = plt.plot(x,[np.mean(y)]*10,color='red',linewidth=2)
std, = plt.plot(x,[np.std(y)]*10,color='black')

mean_legend = plt.legend([mean, original,std],['Mean', 'Random','Standard deviation'])

# Add the legend manually to the current Axes.
ax = plt.gca()
plt.show()
```

shape -  1-D Тензор или python массив. Задаёт форму получаемого тензора.

mean - 0-D Тензор или значение типа dtype.Среднее значение нормального распределения.

stddev - 0-D Тензор или значение типа dtype. Стандартное отклонение нормального распределения.

dtype - Тип выходных значений.

seed -  Целое значение. Используется для создания не повторяемой/повторяемой последовательности для распределения.Смотреть set_random_seed для детального пояснения.

name - опциональное имя операции графа.
